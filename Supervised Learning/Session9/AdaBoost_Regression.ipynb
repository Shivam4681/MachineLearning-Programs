```python
import pandas as pd
```

```python
df1=pd.read_csv("cars.csv")
df1
```

```python
df=df1[['Price','Weight','Horsepower','EngineSize','Rev.per.mile','Wheelbase','Man.trans.avail']]
df
```

```python
df.isna().sum()
```

## Regression

```python
# Define Feature and Target

x=df.drop(['Weight','Man.trans.avail'],axis=1)
y=df['Weight']
```

```python
x
```

```python
y
```

```python
df[['Weight','Price','Horsepower','EngineSize','Rev.per.mile']].corr()
```

### Preprocessing

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x1=pd.DataFrame(ss.fit_transform(x),columns=ss.get_feature_names_out())
x1
```

### Split

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.25,random_state=17)
```

```python
x_train
```

```python
y_train
```

### Model

```python
from sklearn.ensemble import AdaBoostRegressor, AdaBoostClassifier
```

```python
from sklearn.tree import DecisionTreeRegressor,DecisionTreeClassifier
```

```python
dtr=DecisionTreeRegressor(max_depth=2)
dtr
```

```python
abr=AdaBoostRegressor(dtr,random_state=23)
abr
```

### Evaluation

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

### Training Data evaluation

```python
# Fitting

abr.fit(x_train,y_train)
```

```python
# Prediction

y_pred_train=abr.predict(x_train)
```

```python
mse=mean_squared_error(y_pred_train,y_train)
mae=mean_absolute_error(y_pred_train,y_train)
rmse=mse**0.5
r2_score_train=r2_score(y_pred_train,y_train)
```

```python
print("mse=",mse)
print("mae=",mae)
print("rmse=",rmse)
print("R2 Score=",r2_score_train)
```

### Testing Data Evaluation

```python
y_pred_test=abr.predict(x_test)
```

```python
mse=mean_squared_error(y_pred_test,y_test)
mae=mean_absolute_error(y_pred_test,y_test)
rmse=mse**0.5
r2_score_test=r2_score(y_pred_test,y_test)
```

```python
print("Mean Squared Error",mse)
print("Mean Absolute Error",mae)
print("Root Mean Squared Error",rmse)
print("R2_score",r2_score_test)
```

### Hyperparameter Tunning

```python
grid={'n_estimators':range(1,200)}
```

```python
rs=RandomizedSearchCV(abr,param_distributions=grid,cv=3)
rs
```

```python
rs.fit(x_train,y_train)
```

```python
rs.best_params_
```

```python
abr1=rs.best_estimator_
abr1
```

### Feature Importance

```python
col_imp=abr1.feature_importances_
col_imp
```

# ADABOOST(Adaptive Boosting) Classification Problem

```python
# Target 

y = df['Man.trans.avail']
y
```

```python
# Label Encoder

from sklearn.preprocessing import LabelEncoder
la = LabelEncoder()
```

```python
y1 = la.fit_transform(y)
```

```python
y1
```

### Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.2,random_state=23)
```

```python
x_train
```

```python
y_train
```

```python
dtc = DecisionTreeClassifier(max_depth=2)
dtc
```

```python
abc = AdaBoostClassifier(dtc, random_state = 23)
abc
```

### Fitting The Model

```python
abc.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

### Training Data Evaluation

```python
y_pred_train=abc.predict(x_train)
```

```python
clf=classification_report(y_pred_train,y_train)
cnf=confusion_matrix(y_pred_train,y_train)
train_acc=accuracy_score(y_pred_train,y_train)
print("Classfication Report=",clf)
print("Confusion Matrix=",cnf)
print("Accuracy Score=",train_acc)
```

### Testing data 

```python
y_pred_test=abc.predict(x_test)
```

```python
clf=classification_report(y_pred_test,y_test)
cnf=confusion_matrix(y_pred_test,y_test)
test_acc=accuracy_score(y_pred_test,y_test)
print("Classfication Report=",clf)
print("Confusion Matrix=",cnf)
print("Accuracy Score=",test_acc)
```

### Hyperparameter Tunning

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
grid={'n_estimators':range(1,200)}
grid
```

```python
rsc=RandomizedSearchCV(abc,param_distributions=grid,cv=4)
rsc
```

```python
# fit
rsc.fit(x_train,y_train)
```

```python
rsc.best_params_
```

```python
abc1=rsc.best_estimator_
```

```python
abc1
```

### Training Data

```python
abc1.fit(x_train,y_train)
```

```python
y_pred_train=abc1.predict(x_train)
```

```python
clf=classification_report(y_pred_train,y_train)
cnf=confusion_matrix(y_pred_train,y_train)
train_acc=accuracy_score(y_pred_train,y_train)
print("Classfication Report=",clf)
print("Confusion Matrix=",cnf)
print("Accuracy Score=",train_acc)
```

### Testing data

```python
y_pred_test=abc1.predict(x_test)
```

```python
clf=classification_report(y_pred_test,y_test)
cnf=confusion_matrix(y_pred_test,y_test)
test_acc=accuracy_score(y_pred_test,y_test)
print("Classfication Report=",clf)
print("Confusion Matrix=",cnf)
print("Accuracy Score=",test_acc)
```
