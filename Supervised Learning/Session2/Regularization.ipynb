# Preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer # treatment of missing 
from sklearn.preprocessing import StandardScaler,OrdinalEncoder  # scaling issue, categorical issue
from sklearn.compose import ColumnTransformer 
```

```python
# separate cat and con create cat and con list
cat=[]
con=[]

for i in x.columns:
    if x[i].dtypes=='object':
        cat.append(i)
    else:
        con.append(i)
    
```

```python
# display categorical
cat
```

```python
# display continuous
con
```

## What is Ordinal Encoding
Ordinal encoding is a technique that is used to transform categorical variables into a numerical format by assigning a unique value to each of its categories.
It is also referred to as Label Encoding.

# Create PipeLine

# Numerical Pipeline

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='median')),('scalar',StandardScaler())])
```

```python
num_pipe
```

### Categorical Pipeline

```python
cat_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),('encode',OrdinalEncoder())])
cat_pipe
```

```python
# create preprocessing using ColumnTransformer

pre=ColumnTransformer([('cat_pipe',cat_pipe,cat),('num_pipe',num_pipe,con)])
```

```python
pre
```

```python
x1=pd.DataFrame(pre.fit_transform(x),columns=pre.get_feature_names_out())
x1
```

# Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.2,random_state=23)
```

```python
y_train
```

# Model Building

```python
from sklearn.linear_model import LinearRegression, Lasso,Ridge
```

```python
# Create object of LinearRegression
lr=LinearRegression()

# create object of Lasso and Ridge
la=Lasso()
ra=Ridge()
```

# Evaluation

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
# fitting of model
#LinearRegression.fit(feature,target)

lr.fit(x_train,y_train)
```

```python
# invoke/ generate predicted value

y_pred_train=lr.predict(x_train)
```

```python
y_pred_train
```

# Training Data Evaluation

```python
mse=mean_squared_error(y_pred_train,y_train)
mae=mean_absolute_error(y_pred_train,y_train)
r2_score1=r2_score(y_pred_train,y_train)
rmse=mse**0.5
```

```python
print("Mean_Squared_Error",mse)
print("Mean Absolute Error",mae)
print("Root Mean Squared Error",rmse)
print("R2 Score",r2_score1)
```

# Testing Data Evaluation

```python
y_pred_test=lr.predict(x_test)
```

```python
mse=mean_squared_error(y_pred_test,y_test)
mae=mean_absolute_error(y_pred_test,y_test)
r2_score1=r2_score(y_pred_test,y_test)
rmse=mse**0.5
```

```python
print("Mean_Squared_Error",mse)
print("Mean Absolute Error",mae)
print("Root Mean Squared Error",rmse)
print("R2 Score",r2_score1)
```
