```python
import pandas as pd
df1 = pd.read_csv('cars.csv')
df1
```

### check Info

```python
df1.info()
```

```python
# Create dataframe by subsetting classification columns 

df = df1[['Rev.per.mile', 'Horsepower', 'Price', 'Weight', 'Man.trans.avail', 'RPM']]
df
```

# Regression

### Define Feature and Target

```python
x = df.drop(['Price', 'Man.trans.avail'], axis = 1)
y = df['Price']
```

```python
x
```

```python
y
```

# Preprocessing

```python
from sklearn.preprocessing import StandardScaler
```

```python
# create object of Standard scaler
ss = StandardScaler()
```

```python
# transform feature values

x1 = pd.DataFrame(ss.fit_transform(x), columns = ss.get_feature_names_out())
x1
```

## Split

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
```

```python
x_train,x_test,y_train,y_test= train_test_split(x1,y,test_size=0.2,random_state=23)
```

# Model Building

```python
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
```

```python
# create object of DecisionTreeRegressor

dt=DecisionTreeRegressor(random_state=23)
```

# Fit Model

```python
dt.fit(x_train,y_train)
```

# Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

# Training Data Evaluation

```python
# predict

y_pred_train=dt.predict(x_train)
y_pred_train
```

```python
y_pred_train-y_train
```

```python
mse=mean_squared_error(y_pred_train,y_train)
mae=mean_absolute_error(y_pred_train,y_train)
rmse=mse**0.5
r=r2_score(y_pred_train,y_train)

print('MSE:',mse)
print('RMSE:',rmse)
print('R2:',r)
print('MAE:',mae)
```

# Testing Data Evaluation

```python
y_pred_test=dt.predict(x_test)
y_pred_test
```

```python
y_pred_test - y_test
```

```python
mse1=mean_squared_error(y_pred_test,y_test)
mae1=mean_absolute_error(y_pred_test,y_test)
rmse1=mse1**0.5
r1=r2_score(y_pred_test,y_test)

print('MSE:',mse1)
print('RMSE:',rmse1)
print('R2:',r1)
print('MAE:',mae1)
```

> Model Better Performimg on training data but not on testing data

> Thats why we can use RandomisedSearchCV

> Using Puning it will better perform on testing also

# Pruning / Hyperparameter Tuning

```python
# creation of hyperparameter

grid={
    'max_depth':range(1,200),
    'min_samples_split':range(1,20),
    'min_samples_leaf':range(1,20)
}
```

```python
rs=RandomizedSearchCV(dt,param_distributions=grid,cv=3)
rs
```

# fit model

```python
rs.fit(x_train,y_train)
```

```python
# find best parameter

rs.best_params_
```

```python
# find best estimator

dt1=rs.best_estimator_
dt1
```

# Fit Model

```python
dt1.fit(x_train, y_train)
```

```python
# predict 

y_pred_train = dt1.predict(x_train)
```

### Training Data Evaluation

```python
mse=mean_squared_error(y_pred_train,y_train)
mae=mean_absolute_error(y_pred_train,y_train)
rmse=mse**0.5
r=r2_score(y_pred_train,y_train)

print('MSE:',mse)
print('RMSE:',rmse)
print('R2:',r)
print('MAE:',mae)
```

### Test Data Evaluation

```python
y_pred_test = dt1.predict(x_test)
```

```python
mse1=mean_squared_error(y_pred_test,y_test)
mae1=mean_absolute_error(y_pred_test,y_test)
rmse1=mse1**0.5
r1=r2_score(y_pred_test,y_test)

print('MSE:',mse1)
print('RMSE:',rmse1)
print('R2:',r1)
print('MAE:',mae1)
```

## Model

## Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
acc=accuracy_score(y_pred_train,y_train)
clf=classification_report(y_pred_train,y_train)
cnf=confusion_matrix(y_pred_train,y_train)

print('Accuracy:',acc, '\n --------------------------------------------------------\n')
print('Classification Report:\n',clf, '\n --------------------------------------------------------\n')
print('Confusion Matrix:\n',cnf)
```

### Testing Data Evalaution

```python
y_pred_test=dtc.predict(x_test)
```

```python
acc1=accuracy_score(y_pred_test,y_test)
clf1=classification_report(y_pred_test,y_test)
cnf1=confusion_matrix(y_pred_test,y_test)

print('Accuracy:',acc1, '\n --------------------------------------------------------\n')
print('Classification Report:\n',clf1, '\n --------------------------------------------------------\n')
print('Confusion Matrix:\n',cnf1)
```
