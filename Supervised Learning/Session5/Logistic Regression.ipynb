### define feature and targets

```python
x = df1.drop(['Man.trans.avail'], axis = 1)
y = df1['Man.trans.avail']
```

# Preprocessing

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# create objects of StandardScaler, LabelEncoder

ss = StandardScaler()
le = LabelEncoder()
```

```python
# transform Feature Variable x

x1 = pd.DataFrame(ss.fit_transform(x), columns = ss.get_feature_names_out())
y1 = le.fit_transform(y)
```

```python
# display x1
x1
```

```python
# display y1
y1
```

# Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x1, y1, test_size = 0.2, random_state = 25)
```

# Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
# create object of LogisticRegression

lr = LogisticRegression()
```

# Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# Training Data Evaluation

```python
# create predicted value

y_pred_train = lr.predict(x_train)
y_pred_train
```

```python
acc = accuracy_score(y_pred_train, y_train)
```

```python
clf = classification_report(y_pred_train, y_train)
```

```python
cnf = confusion_matrix(y_pred_train, y_train)
```

```python
print('Accuracy:',acc, '\n', '-------------------------------------------------------', '\n')
print('Classification Report:\n',clf, '\n', '-------------------------------------------------------', '\n')
print('Confusion Matrix:\n',cnf, '\n', '-------------------------------------------------------', '\n')
```
