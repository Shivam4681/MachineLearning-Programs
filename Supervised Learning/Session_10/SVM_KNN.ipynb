```python
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
df
```

```python
df1 = df[['Price', 'Weight', 'Horsepower', 'EngineSize', 'Rev.per.mile', 'Wheelbase', 'Man.trans.avail']]
df1
```

```python
df1.isna().sum()
```

## Regression

```python
# Define Feature And Target

x = df1.drop(['Weight', 'Man.trans.avail'], axis = 1)
y = df1['Weight']
```

```python
x
```

```python
y
```

### Preprocessing

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
# Transform feature variable

x1=pd.DataFrame(ss.fit_transform(x),columns=ss.get_feature_names_out())
x1
```

### Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x1, y, train_size = 0.8, random_state = 16)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

### Model Selection

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
kn = KNeighborsRegressor()
svr = SVR()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

### Model Fitting  KNN

```python
kn.fit(x_train, y_train)
```

### Evaluation

#### Training Data

```python
y_pred_train = kn.predict(x_train)
```

```python
mse = mean_squared_error(y_pred_train, y_train)
mae = mean_absolute_error(y_pred_train, y_train)
rmse = mse**0.5
r2_score_train = r2_score(y_pred_train, y_train)
```

```python
print('mean_squared_error', mse)
print('mean_absolute_error', mae)
print('Root_mean_squared_error', rmse)
print('r2_score', r2_score_train)
```

#### Testing Data

```python
y_pred_test = kn.predict(x_test)
```

```python
mse=mean_squared_error(y_pred_test,y_test)
mae=mean_absolute_error(y_pred_test,y_test)
rmse=mse**0.5
r2_score_test=r2_score(y_pred_test,y_test)
```

```python
print("Mean Squared Error",mse)
print("Mean Absolute Error",mae)
print("Root Mean Squared Error",rmse)
print("R2_score",r2_score_test)
```

### Model Fitting SVM

```python
svr.fit(x_train, y_train)
```

### Evaluation

#### Training data

```python
y_pred_train = svr.predict(x_train)
```

```python
mse = mean_squared_error(y_pred_train, y_train)
mae = mean_absolute_error(y_pred_train, y_train)
rmse = mse**0.5
r2_score_train = r2_score(y_pred_train, y_train)
```

```python
print('mean_squared_error', mse)
print('mean_absolute_error', mae)
print('Root_mean_squared_error', rmse)
print('r2_score', r2_score_train)
```

```python
y_pred_test = svr.predict(x_test)
```
