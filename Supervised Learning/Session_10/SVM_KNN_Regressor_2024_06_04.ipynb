```python
import pandas as pd
```

```python
df1=pd.read_csv("cars.csv")
df1
```

```python
# Regression SVM, KNN
```

```python
df=df1[['Price','Weight','Horsepower','EngineSize','Rev.per.mile','Wheelbase','Man.trans.avail']]
df
```

```python
df.isna().sum()
```

# Regression
# Define Features. Target

```python
x=df.drop(['Weight','Man.trans.avail'],axis=1)
x
```

```python
# target
y=df['Weight']
```

```python
y
```

# Preprocessing

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
# transform feature variable
x1=pd.DataFrame(ss.fit_transform(x),columns=ss.get_feature_names_out())
```

```python
x1
```

```python
# Splitting
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y,train_size=0.8,random_state=16)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Model
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

# create objects


```python
kn=KNeighborsRegressor()
svr=SVR()
```

# Fitting Model

```python
kn.fit(x_train,y_train)
```

# Evaluation

```python
# Training Data Evaluation

y_pred_train=kn.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse=mean_squared_error(y_pred_train,y_train)
mae=mean_absolute_error(y_pred_train,y_train)
rmse=mse**0.5
r2_score_train=r2_score(y_pred_train,y_train)
print("Mean Squared Error",mse)
print("Mean Absolute Error",mae)
print("Root Mean Squared Error",rmse)
print("R2_score",r2_score_train)
```

```python
# Testing data
```

```python
y_pred_test=kn.predict(x_test)
```

```python
mse=mean_squared_error(y_pred_test,y_test)
mae=mean_absolute_error(y_pred_test,y_test)
rmse=mse**0.5
r2_score_test=r2_score(y_pred_test,y_test)
print("Mean Squared Error",mse)
print("Mean Absolute Error",mae)
print("Root Mean Squared Error",rmse)
print("R2_score",r2_score_test)
```
