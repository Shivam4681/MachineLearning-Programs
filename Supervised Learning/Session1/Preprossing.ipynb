> Missing Data Handling

```python
df.isna().sum()
```

```python
# treating missing values
for i in df.columns:
    if df[i].dtypes == 'object':
        m = df[i].mode()[0]
        df[i] = df[i].fillna(m)
    else:
        ma = df[i].mean()
        df[i] = df[i].fillna(ma)
```

```python
# check for missing values
df.isna().sum()
```

### Define Target

```python
# indipendent variable

x = df.drop(['Price'], axis = 1)    # feature variable

# dependent variable

y = df['Price']      # Target variable
```

### Preprocessing

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# create object of LabelEncoder
le = LabelEncoder()

# create object of OneHotEncoder
oh = OneHotEncoder()
```

## Split

```python
from sklearn.model_selection import train_test_split
```

> generating training and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x1,y,test_size = 0.2, random_state = 30)
```

## Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
# create object of Linear Regression
la = LinearRegression()
```

```python
# fit model
# linearRegression.fit(feature_train,target_train)
la.fit(x_train, y_train)
```

```python
#check intercept
la.intercept_
```

```python
# check coefficient
la.coef_
```

## Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

> Training Data Evaluation

```python
# generate prediction
y_pred_train = la.predict(x_train)
```

```python
y_pred_train
```

```python
y_train
```

```python
# Error
y_train-y_pred_train
```

```python
# mean_squared_error(actual,predicted)
mse = mean_squared_error(y_train, y_pred_train)
mae = mean_absolute_error(y_train, y_pred_train)
r2_score1 = r2_score(y_train, y_pred_train)
rmse = mse**0.5
```

```python
print("Mean_Squared_Error",mse)
print("Mean Absolute Error",mae)
print("Root Mean Squared Error",rmse)
print("R2 Score",r2_score1)
```
