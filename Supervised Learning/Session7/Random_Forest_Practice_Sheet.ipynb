```python
# create sub set of df consisting of Price,Weight,RPM,Rev.per.mile,Horsepower,EngineSize,Wheelbase,Man.tran.avail
```

# Regression

```python
# define feature  and target is price

x = df1.drop(['Price', 'Man.trans.avail'], axis = 1)
y = df1['Price']
```

```python
x
```

```python
y
```

# Preprocessing

```python
from sklearn.preprocessing import StandardScaler
```

```python
# create object of Standard Scaler

ss = StandardScaler()
```

```python
# transform feature variable

x1 = pd.DataFrame(ss.fit_transform(x), columns = ss.get_feature_names_out())
x1
```

# Split

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.2,random_state=23)
```

```python
x_train
```

```python
y_train
```

```python
y_test
```

# Model

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
```

```python
rf = RandomForestRegressor(random_state = 23)
```

# fit model

```python
rf.fit(x_train, y_train)
```

# Evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

### Training Data Evaluation

```python
y_pred_train = rf.predict(x_train)
```

```python
mse=mean_squared_error(y_pred_train,y_train)
mae=mean_absolute_error(y_pred_train,y_train)
rmse=mse**0.5
r2_score_train=r2_score(y_pred_train,y_train)
```

```python
print("mse:",mse)
print("mae:",mae)
print("rmse:",rmse)
print("R2 Score:",r2_score_train)
```

### Testing Data Evaluation

```python
y_pred_test = rf.predict(x_test)
```

```python
mse=mean_squared_error(y_pred_test,y_test)
mae=mean_absolute_error(y_pred_test,y_test)
rmse=mse**0.5
r2_score_test=r2_score(y_pred_test,y_test)
```

```python
print("mse:",mse)
print("mae:",mae)
print("rmse:",rmse)
print("R2 Score:",r2_score_test)
```

# Hyperparameter Tuning

```python
import numpy as np
grid={
    'n_estimators':range(1,300),
    'max_depth':range(1,300),
    'min_samples_split':range(1,20),
    'min_samples_leaf':range(1,20),
    'max_samples':np.arange(0.1,1,0.1)
    
}
```

```python
rs=RandomizedSearchCV(rf,param_distributions=grid,cv=3)
rs
```

# Model

```python
rfc=RandomForestClassifier(random_state=23)
```

```python
# fit model

rfc.fit(x_train,y_train)
```

# Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

### Training Data Evaluation

```python
y_pred_train=rfc.predict(x_train)
```

```python
acc_score=accuracy_score(y_pred_train,y_train)
clf=classification_report(y_pred_train,y_train)
cnf=confusion_matrix(y_pred_train,y_train)
```

```python
print("Accuracy Score=",acc_score, '\n----------------------------------------------------\n')
print("Classification report=",clf, '\n----------------------------------------------------\n')
print("Confusion_Matrix=",cnf, '\n----------------------------------------------------\n')
```

### Testining Data Evaluation

```python
y_pred_test=rfc.predict(x_test)
```

```python
acc_score=accuracy_score(y_pred_test,y_test)
clf=classification_report(y_pred_test,y_test)
cnf=confusion_matrix(y_pred_test,y_test)
```

```python
print("Accuracy Score=",acc_score, '\n----------------------------------------------------\n')
print("Classification report=",clf, '\n----------------------------------------------------\n')
print("Confusion_Matrix=",cnf, '\n----------------------------------------------------\n')
```
