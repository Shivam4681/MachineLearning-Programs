## Seggreagation of columns

```python
cat=[]
con=[]

for i in x.columns:
    if x[i].dtypes == 'object':
        cat.append(i)
    else:
        con.append(i)
```

# Preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

## Create PipeLine

## Numerical Pipeline

```python
# num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='median')),('scalar',StandardScaler())])

num_pipe = Pipeline(steps = [('impute', SimpleImputer(strategy = 'median')),('scalar', StandardScaler())])
```

```python
num_pipe
```

## Categorical Pipeline

```python
# cat_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),('encode',OrdinalEncoder())])

cat_pipe = Pipeline(steps = [('impute', SimpleImputer(strategy = 'most_frequent')), ('encode', OrdinalEncoder())])
```

```python
cat_pipe
```

```python
# Column Transformer
# create preprocessing using ColumnTransformer

pre = ColumnTransformer([('cat', cat_pipe, cat), ('con', num_pipe, con)])
```

```python
pre
```

```python
# Fit Trnasform
x1 = pd.DataFrame(pre.fit_transform(x), columns=pre.get_feature_names_out())
```

```python
x1
```

# ANOVA

```python
A = df[['Type', 'Weight']]
A
```

```python
from statsmodels.formula.api import ols
```

```python
model = ols('Weight ~ Type', A).fit()
```

```python
from statsmodels.stats.anova import anova_lm
```

```python
model1 = anova_lm(model)
model1
```

```python
p = model1.iloc[0:1,4:5]
p
```

```python
# check p value

p_val=p.values[0][0]
p_val
```

```python
# checking the hypothesis

if p_val>0.05:
    print('Null Hypothesis')
    print('No Relation between column')
else:
    print('ALternate Hypothesis')
    print('Columns are related')
```

# Chi square Test

```python
B = df[['AirBags', 'Type']]
B
```

```python
ctab  = pd.crosstab(B['AirBags'], B['Type'])
```

```python
from scipy.stats import chi2_contingency
```

```python
a,b,c,d=chi2_contingency(ctab)
```

```python
a
```

```python
b
```

```python
if b>0.05:
    print('Null Hypothesis')
    print('No Association between column')
else:
    print('ALternate Hypothesis')
    print('Association Between Columns')
```
